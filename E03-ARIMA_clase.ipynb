```python
plot_pacf(df, lags=37);
```

```python
from statsmodels.tsa.arima_model import ARIMA
# fit model

X = df.values
size = int(len(X) * 0.5)
train, test = X[0:size], X[size:len(X)]
predictions = list()
errors = list()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for i in range(5):
    for j in range(5):
        for k in range(5):
            try:
                for t in range(len(test)):
                    model = ARIMA(X, order=(i,1,k))
                    model_fit = model.fit(disp=0)
                    output = model_fit.forecast()
                    yhat = output[0]
                    predictions.append(yhat)
                    error = mean_squared_error(test[t], yhat)**0.5
                    errors.append(error)
                    print("P:" + str(i) + " D:" + str(j) + " Q:" + str(k))
                    print('predicted=%f, expected=%f' + str(yhat))
                    print("MSE" + str(error))
            except:
                pass
```

```python
for i in range(3):
    for j in range(3):
        for k in range(3):
            model = ARIMA(X, order=(i,j,k))
            model_fit = model.fit(disp=0)
            output = model_fit.forecast()
            yhat = output[0]
            predictions.append(yhat)
            print("P:" + str(i) + " D:" + str(j) + " Q:" + str(k))
            print('predicted=%f, expected=%f' + str(yhat))
```
